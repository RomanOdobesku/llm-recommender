```python
from typing import Tuple


def get_predictions(cat_1: str, cat_2: str) -> Tuple[str, str]:
    """
    Predicts next 5 categories for user based on 2 previous ones
    """
    content = f"User's previous categories: {cat_1}, {cat_2}"

    completion = client.beta.chat.completions.parse(
        model="gpt-4o-2024-08-06",
        messages=[
            {"role": "system", "content": prompt},
            {"role": "user", "content": f"{content}"},
        ],
        response_format=Response,
    )
    message = completion.choices[0].message

    print(message)

    predicted_categories = []
    for category in message.parsed.categories:
        predicted_categories.append(
            (category.dict()["category"].value, category.dict()["reason"])
        )

    return predicted_categories
```

```python
columns = ["user_category_1", "user_category_2", "predicted_category", "explanation"]
recommendation_table = pd.DataFrame(columns=columns)


# Функция для добавления данных в таблицу
def add_user_categories_and_predictions(user_category1, user_category2, predictions):
    global recommendation_table

    # Добавляем данные
    for category, explanation in predictions:
        recommendation_table = pd.concat(
            [
                recommendation_table,
                pd.DataFrame(
                    {
                        "user_category_1": [user_category1],
                        "user_category_2": [user_category2],
                        "predicted_category": [category],
                        "explanation": [explanation],
                    }
                ),
            ],
            ignore_index=True,
        )


cat_1 = "Контейнеры для еды"
cat_2 = "Купальники"
predicted_categories = get_predictions(cat_1, cat_2)

# Добавляем данные в таблицу
add_user_categories_and_predictions(cat_1, cat_2, predicted_categories)
```
