```python
import warnings
from enum import Enum
from typing import List

import pandas as pd
from dotenv import load_dotenv
from openai import OpenAI
from pydantic import BaseModel

warnings.filterwarnings("ignore")

load_dotenv(override=True)


class CategoryEnum(str, Enum):
    smartphones = "Смартфоны"
    accessories = "Аксессуары для смартфонов и телефонов"
    gaming_genres = "Игровые жанры"
    playstation = "PlayStation"
    nintendo = "Nintendo"
    xbox = "Xbox"
    gaming_consoles = "Игровые приставки"
    pc = "PC"
    mobile_gaming = "Mobile gaming"
    retro_consoles = "Ретро-консоли"
    gaming_cartridges = "Игровые картриджи"
    gaming_merchandise = "Игровая атрибутика"
    gaming_headsets = "Игровые наушники"
    gamepads = "Геймпады"
    vr_glasses = "Очки виртуальной реальности"
    wheels_and_pedals = "Рули и педали"
    console_accessories = "Аксессуары для игровых приставок"
    esports_accessories = "Аксессуары для киберспорта"
    batteries_and_chargers = "Аккумуляторы и зарядки"
    bags_and_cases = "Сумки и чехлы"
    console_parts = "Запчасти для игровых консолей"
    phones_and_smartwatches = "Телефоны и смарт-часы"
    headphones_and_audio = "Наушники и аудиотехника"
    laptops_and_tablets = "Ноутбуки, планшеты и электронные книги"
    tvs_and_video_equipment = "Телевизоры и видеотехника"
    pc_components = "Комплектующие для ПК"
    computers_and_peripherals = "Компьютеры и периферия"
    gaming_laptops_and_consoles = "Игровые приставки и ноутбуки"
    office_equipment = "Офисная техника"
    cameras_and_video_cameras = "Фото и видеокамеры"
    security_systems = "Охранные системы и видеонаблюдение"
    smart_home = "Умный дом"
    drones_and_accessories = "Квадрокоптеры и аксессуары"
    navigators = "Навигаторы"
    optical_devices = "Оптические приборы"
    clocks_and_alarms = "Часы и электронные будильники"
    electronics_accessories = "Аксессуары для электроники"
    garden_furniture = "Садовая мебель"
    computer_and_office_furniture = "Компьютерная и офисная мебель"
    soft_furniture = "Мягкая мебель"
    tables_and_chairs = "Столы и стулья"
    storage_furniture = "Мебель для хранения"
    bedroom_furniture = "Мебель для спальни и комплектующие"
    kitchen_furniture = "Мебель для кухни"
    bathroom_furniture = "Мебель для ванной"
    frameless_furniture = "Бескаркасная мебель"
    childrens_furniture = "Детская мебель"
    business_furniture = "Мебель для бизнеса"
    safes_and_archive_cabinets = "Сейфы и архивные шкафы"
    music = "Музыка"
    vinyl_records = "Виниловые пластинки"
    video = "Видео"
    new_cars = "Автомобили новые"
    school_and_learning_goods = "Товары для школы и обучения"
    toys_and_games = "Игрушки и игры"
    sports_and_outdoor_games = "Спорт и игры на улице"
    diapers_and_hygiene = "Подгузники и гигиена"
    childrens_room = "Детская комната"
    strollers_and_car_seats = "Коляски и автокресла"
    baby_food = "Детское питание"
    feeding_products = "Товары для кормления"
    products_for_moms = "Товары для мам"
    writing_supplies = "Письменные принадлежности"
    paper = "Бумага"
    paper_products = "Бумажная продукция"
    folders_and_files = "Папки и файлы"
    covers = "Обложки"
    display_boards = "Демонстрационные доски"
    office_supplies = "Офисные принадлежности"
    stands_and_card_holders = "Подставки и визитницы"
    seals_and_stamps = "Печати и штампы"
    calculators = "Калькуляторы"
    laser_printer_cartridges = "Картриджи для лазерных принтеров"
    inkjet_printer_cartridges = "Картриджи для струйных принтеров"
    drawing_tools = "Чертежные принадлежности"
    trading_equipment = "Оборудование и принадлежности для торговли"
    printing_equipment = "Полиграфическое оборудование"
    sublimation_printing = "Сублимационная печать"
    tableware_and_kitchen_utensils = "Посуда и кухонные принадлежности"
    textiles = "Текстиль"
    lighting = "Освещение"
    decor_and_interior = "Декор и интерьер"
    gardening = "Дача и сад"
    flowers_and_pots = "Цветы, растения и горшки"
    household_goods = "Хозяйственные товары"
    storage = "Хранение вещей"
    sauna_goods = "Товары для бань и саун"
    holiday_goods = "Товары для праздников"
    religion_and_esoterics = "Религия и эзотерика"
    ritual_goods = "Ритуальные товары"
    women = "Женщинам"
    men = "Мужчинам"
    children = "Детям"
    care_and_accessories = "Уход и аксессуары"
    rings = "Кольца"
    earrings = "Серьги"
    bracelets = "Браслеты"
    necklaces = "Украшения на шею"
    jewelry_sets = "Комплекты украшений"
    charms = "Шармы"
    brooches_and_pins = "Броши и значки"
    piercing = "Пирсинг"
    watches = "Часы"
    childrens_jewelry = "Детские украшения"
    cufflinks_and_clips = "Запонки и зажимы"
    souvenirs = "Сувениры"
    laces = "Шнурки"
    hair_accessories = "Аксессуары для волос"
    religious_jewelry = "Религиозные украшения"
    electronic_certificates = "Электронные сертификаты"
    online_subscriptions = "Онлайн-подписки"
    games_and_subscriptions = "Игры и игровые подписки"
    software = "Программное обеспечение"
    cloud_storage = "Облачные хранилища"
    services = "Услуги"
    tourism_and_outdoors = "Туризм и отдых на природе"
    fishing = "Рыбалка"
    hunting_and_shooting = "Охота и стрельба"
    boats_and_motors = "Лодки и лодочные моторы"
    fishing_and_hunting_clothing = "Одежда для рыбалки и охоты"
    fishing_and_hunting_footwear = "Обувь для рыбалки и охоты"
    tactical_clothing = "Тактическая одежда и аксессуары"
    survival_kits = "Наборы для выживания"
    nonfiction = "Нехудожественная литература"
    business_books = "Бизнес-литература"
    fiction = "Художественная литература"
    children_and_parents = "Детям и родителям"
    educational_books = "Учебная литература"
    e_books = "Электронные книги"
    rare_books = "Букинистика"
    comics = "Комиксы"
    manga = "Манга"
    on_demand_printing = "Печать по требованию"
    gift_editions = "Подарочные издания книг"
    magazines_and_newspapers = "Журналы и газеты"
    foreign_language_books = "Литература на иностранных языках"
    audiobooks = "Электронные аудиокниги"
    ebook_and_audiobook_subscriptions = "Подписки на электронные книги и аудиокниги"
    antique_books = "Антикварные книги"
    used_books = "Книги б/у"
    notebooks_and_diaries = "Блокноты и ежедневники"
    booklets_and_brochures = "Буклеты, листовки и брошюры"
    travel = "Путешествия"
    hair_care = "Уход за волосами"
    face_care = "Уход за лицом"
    body_care = "Уход за телом"
    makeup = "Макияж"
    manicure_and_pedicure = "Маникюр и педикюр"
    perfumery = "Парфюмерия"
    tanning_and_sun_protection = "Загар и защита от солнца"
    mens_cosmetics = "Мужская косметика"
    childrens_cosmetics_and_perfumes = "Детская косметика и парфюмерия"
    for_pregnant_and_nursing = "Беременным и кормящим"
    hardware_cosmetology = "Аппаратная косметология и массаж"
    aromatherapy = "Ароматерапия"
    tattoo_salon_supplies = "Оборудование и материалы для тату-салона"
    beauty_salon_equipment = "Мебель и оборудование для салонов красоты"
    medicines = "Лекарственные средства"
    vitamins_and_supplements = "Витамины и БАДы"
    para_pharmacy = "Парафармацевтика"
    optics = "Оптика"
    medical_devices = "Медицинские приборы"
    medical_equipment = "Медицинское оборудование"
    accessories_and_parts = "Аксессуары и комплектующие"
    medical_supplies = "Медицинские изделия и расходные материалы"
    medical_clothing_and_textiles = "Одежда и текстиль медицинские"
    medical_tools = "Медицинские инструменты"
    dental_materials = "Материалы и средства стоматологические"
    hygiene_products = "Товары для гигиены"
    rehabilitation_goods = "Товары для реабилитации"
    orthopedics = "Ортопедия"
    hand_massagers = "Ручные массажеры и аппликаторы"
    medical_furniture = "Мебель медицинская"
    uniforms_and_workwear = "Униформа и рабочая одежда"
    clothing_care = "Уход за одеждой"
    climate_technology = "Климатическая техника"
    kitchen_appliances = "Техника для кухни"
    home_appliances = "Техника для дома"
    beauty_and_health_technology = "Техника для красоты и здоровья"
    large_appliances = "Крупная бытовая техника"
    catering_equipment = "Техника для общепита"
    collectibles = "Коллекционирование"
    tableware = "Посуда и утварь"
    military_history_items = "Предметы военной истории"
    interior_items = "Предметы интерьера"
    household_items = "Предметы быта"
    furniture = "Мебель"
    jewelry = "Украшения"
    wristwatches = "Часы наручные"
    stationery = "Канцелярия"
    instruments = "Приборы и инструменты"
    paintings_and_graphics = "Живопись и графика"
    haberdashery = "Галантерея"
    smoking_accessories = "Товары для курения"
    musical_instruments = "Музыкальные инструменты"
    religious_attribution = "Религиозная атрибутика"
    car_parts = "Запчасти для легковых автомобилей"
    car_accessories = "Автоаксессуары и принадлежности"
    oils_and_auto_chemicals = "Масла и автохимия"
    tires_and_rims = "Шины и диски"
    car_audio = "Автозвук"
    vehicles = "Транспортные средства"
    car_care = "Уход за автомобилем"
    car_electronics = "Электроника для автомобиля"
    tools_and_equipment = "Инструменты и оборудование"
    garage_and_autoservice = "Гараж и автосервис"
    batteries_and_accessories = "Аккумуляторы и аксессуары"
    moto_goods = "Мототовары"
    truck_parts = "Запчасти для грузовиков и спецтехники"
    car_literature = "Автолитература"
    for_cats = "Для кошек"
    for_dogs = "Для собак"
    for_rodents = "Для грызунов и хорьков"
    for_birds = "Для птиц"
    for_fish_and_reptiles = "Для рыб и рептилий"
    for_horses = "Для лошадей"
    veterinary_pharmacy = "Ветаптека"
    farming_goods = "Фермерское хозяйство"
    veterinary_equipment = "Ветеринарное оборудование"
    household_chemicals = "Бытовая химия"
    personal_hygiene = "Личная гигиена"


class CategoryPredict(BaseModel):
    category: CategoryEnum
    reason: str


class Response(BaseModel):
    categories: List[CategoryPredict]


client = OpenAI()

prompt = """
You are an expert in e-commerce and marketplace analytics. Your task is to analyze user preferences based on their past category choices and predict the next product categories they are likely to prefer. The user has expressed interest in two specific product categories. Your goal is to predict exactly 5 new categories they might like, focusing on exploratory recommendations while drawing connections to both of the user's past preferences.

For each prediction, provide the following details:
- **category**: The predicted product category. This should propose diverse and exploratory suggestions that expand on the user's potential interests based on general knowledge and contextual assumptions (e.g., age group, lifestyle, family composition, hobbies, or broader needs).
- **reason**: A concise and logical explanation for why this category was predicted. Each prediction should incorporate connections to both of the user's past preferences, either directly or through creative, plausible inferences. The explanation should be really short and be fitted in 1 sentence.

Ensure that:
- All predictions are exploratory and introduce fresh perspectives on what the user might find interesting or useful.
- Each prediction is based on a hypothesis that takes into account both of the user's past category choices, finding a meaningful link or bridging their potential needs.
- The reasoning is varied and demonstrates creativity, reflecting diverse aspects of the user's potential context.
"""
```

```python
from typing import Tuple


def get_predictions(cat_1: str, cat_2: str) -> Tuple[str, str]:
    """
    Predicts next 5 categories for user based on 2 previous ones
    """
    content = f"User's previous categories: {cat_1}, {cat_2}"

    completion = client.beta.chat.completions.parse(
        model="gpt-4o-2024-08-06",
        messages=[
            {"role": "system", "content": prompt},
            {"role": "user", "content": f"{content}"},
        ],
        response_format=Response,
    )
    message = completion.choices[0].message

    predicted_categories = []
    for category in message.parsed.categories:
        predicted_categories.append(
            (category.dict()["category"].value, category.dict()["reason"])
        )

    return predicted_categories
```

```python
columns = ["user_category_1", "user_category_2", "predicted_category", "explanation"]
recommendation_table = pd.DataFrame(columns=columns)


# Функция для добавления данных в таблицу
def add_user_categories_and_predictions(user_category1, user_category2, predictions):
    global recommendation_table

    # Добавляем данные
    for category, explanation in predictions:
        recommendation_table = pd.concat(
            [
                recommendation_table,
                pd.DataFrame(
                    {
                        "user_category_1": [user_category1],
                        "user_category_2": [user_category2],
                        "predicted_category": [category],
                        "explanation": [explanation],
                    }
                ),
            ],
            ignore_index=True,
        )
```

```python
from itertools import combinations

from tqdm import tqdm

category_pairs = list(combinations(CategoryEnum, 2))
# Определяем шаги для промежуточного сохранения
save_steps = len(category_pairs) // 5  # Каждые 20%

for idx, pair in enumerate(tqdm(category_pairs, desc="Processing category pairs")):
    user_cat_1 = pair[0].value
    user_cat_2 = pair[1].value

    predicted_categories = get_predictions(user_cat_1, user_cat_2)
    add_user_categories_and_predictions(user_cat_1, user_cat_2, predicted_categories)

    # Сохраняем таблицу каждые 25% прогресса
    if (idx + 1) % save_steps == 0 or (idx + 1) == len(
        category_pairs
    ):  # Также сохраняем на последнем шаге
        progress_percent = ((idx + 1) / len(category_pairs)) * 100
        file_name = f"../data/new_categories_{int(progress_percent)}_percent.csv"
        recommendation_table.to_csv(file_name, index=False)
        print(f"Промежуточная таблица сохранена в файл: {file_name}")

recommendation_table.to_csv("../data/new_categories.csv", index=False)
```
