```python
import json
import random

def split_jsonl(input_file, output_file1, output_file2, ratio=0.8):
    """
    Разделить JSONL файл на два файла с заданным соотношением.

    :param input_file: Входной JSONL файл.
    :param output_file1: Выходной JSONL файл для 80% записей.
    :param output_file2: Выходной JSONL файл для 20% записей.
    :param ratio: Соотношение записей в первом файле (по умолчанию 0.8).
    """
    with open(input_file, 'r') as f_in:
        data = [json.loads(line) for line in f_in]

    random.shuffle(data)

    split_index = int(len(data) * ratio)

    with open(output_file1, 'w') as f_out1:
        for item in data[:split_index]:
            f_out1.write(json.dumps(item,ensure_ascii=False) + '\n')

    with open(output_file2, 'w') as f_out2:
        for item in data[split_index:]:
            f_out2.write(json.dumps(item,ensure_ascii=False) + '\n')

# Пример использования
split_jsonl('json_dataset.jsonl', 'train.jsonl', 'test.jsonl')
```

```python
import random
l = pd.read_csv("category_triplets.csv")
csvs = ["new_categories_34_percent.csv","new_categories.csv","category_triplets.csv","sampled_triplets_with_rand_cats.csv"]

jsonl_filename = "json_dataset.jsonl"
with open(jsonl_filename, 'w') as jsonl_file:

    # Создаем csv reader
    for cs in csvs:
        csv = pd.read_csv(cs)
        #csv[csv['predicted_category'].isin(csv["user_category_2"].unique())]
        csv = csv.groupby(["user_category_1","user_category_2"])["predicted_category"].apply(list)
        csv = csv.reset_index(name='predicted_category')

        order = list(range(len(csv)))
        random.shuffle(order)
        # Итерируемся по строкам csv файла
        for i in order:
            row = csv.iloc[i]
            #print(row)
            # Преобразуем строку в json объект
            k = [
                {"role": "system", "content": \
                 """Ты T-lite, виртуальный ассистент в Т-Технологии. Твоя задача - рекомендовать пользователям следующие категории, которые могут им понравиться.
                    Пользователи проявляют интерес к игровым жанрам и сувенирам, что указывает на увлечение развлечениями. Высокая вероятность лайка для новых автомобилей свидетельствует об интересе к крупным и качественным покупкам, вызывающим вау-эффект. Они ценят комфорт, что отражается в интересе к мягкой мебели и текстилю. Забота о здоровье проявляется в интересе к витаминам и БАДам, а также технике для красоты. Пользователи активно интересуются электроникой, включая игровые приставки и мобильные игры, и аксессуары, такие как рули с педалямми.Также пользователям нравятся товары для дома и личной гигиены, что подчеркивает их стремление к уюту и комфорту."""},
                {"role": "user", "content": "Какие три следующие категории понравятся пользователю, если до этого ему понравились следующие категории: "+\
                 row["user_category_1"]+ ", " + row["user_category_2"]},
                { "role" :  "assistant" ,  "content" :  "\n".join(row["predicted_category"][:3]) }
            ]
            json_obj = json.dumps({"messages":k},ensure_ascii=False)

            # Записываем json объект в jsonl файл
            jsonl_file.write(json_obj + '\n')

print(f'Jsonl файл создан: {jsonl_filename}')
```
